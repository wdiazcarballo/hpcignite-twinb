```python
import subprocess
import os
import shutil

# Setup directories
base_dir = os.getcwd()
output_dir = os.path.join(base_dir, "output")

if os.path.exists(output_dir):
    shutil.rmtree(output_dir)
os.makedirs(output_dir, exist_ok=True)

# Copy files
idf_source = "Boonchoo_Building_test.idf"
epw_source = "THA_NRG_Lampang.Agromet.483340_TMYx.2009-2023.epw"

shutil.copy(idf_source, os.path.join(output_dir, "in.idf"))
shutil.copy(epw_source, os.path.join(output_dir, "in.epw"))

eplus_dir = "/project/lt200291-ignite/Project_chomwong/energyplus/EnergyPlus-25.1.0-1c11a3d85f-Linux-CentOS7.9.2009-x86_64"

# หาไฟล์ Energy+.idd
idd_file = os.path.join(eplus_dir, "Energy+.idd")
print(f"IDD file exists: {os.path.exists(idd_file)}")

# Copy Energy+.idd ไปที่ output directory
if os.path.exists(idd_file):
    shutil.copy(idd_file, os.path.join(output_dir, "Energy+.idd"))
else:
    print("❌ Cannot find Energy+.idd")
    # ลองหาไฟล์ที่มี .idd
    for f in os.listdir(eplus_dir):
        if f.endswith('.idd'):
            print(f"Found: {f}")

# เปลี่ยนไปที่ output directory แล้วรัน
os.chdir(output_dir)

# รัน ExpandObjects
expandobjects_exe = os.path.join(eplus_dir, "ExpandObjects")
result = subprocess.run([expandobjects_exe], capture_output=True, text=True)
print("\nExpandObjects output:", result.stdout)

if os.path.exists("expanded.idf"):
    print("✓ ExpandObjects สำเร็จ!")
    
    # รัน EnergyPlus
    energyplus_exe = os.path.join(eplus_dir, "energyplus")
    result = subprocess.run([
        energyplus_exe,
        "-w", "in.epw",
        "-d", ".",
        "expanded.idf"
    ], capture_output=True, text=True)
    print("\nEnergyPlus output:", result.stdout[:500])  # แสดงบางส่วน
else:
    print("❌ ExpandObjects ล้มเหลว")
    if os.path.exists("eplusout.err"):
        with open("eplusout.err", "r") as f:
            print(f.read())

# กลับไป base directory
os.chdir(base_dir)
```

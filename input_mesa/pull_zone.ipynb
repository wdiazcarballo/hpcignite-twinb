```python
# ‡∏î‡∏∂‡∏á‡∏ä‡∏∑‡πà‡∏≠ Zone ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏à‡∏≤‡∏Å‡πÑ‡∏ü‡∏•‡πå .idf
import re 
idf_path = "/project/lt200291-ignite/Project_chomwong/project/EnergyPlus_BP_Boonchoo/output/expanded.idf"  # <-- ‡πÅ‡∏Å‡πâ path ‡∏ï‡∏≤‡∏°‡∏à‡∏£‡∏¥‡∏á

# üîç ‡πÄ‡∏Å‡πá‡∏ö‡∏ä‡∏∑‡πà‡∏≠‡πÇ‡∏ã‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
zone_names = []

# ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå
with open(idf_path, "r", encoding="utf-8", errors="ignore") as f:
    content = f.read()

# ‡πÉ‡∏ä‡πâ regex ‡∏´‡∏≤‡∏ä‡∏∑‡πà‡∏≠ Zone ‡∏à‡∏≤‡∏Å‡∏ö‡∏•‡πá‡∏≠‡∏Å‡∏ó‡∏µ‡πà‡∏Ç‡∏∂‡πâ‡∏ô‡∏ï‡πâ‡∏ô‡∏î‡πâ‡∏ß‡∏¢ "Zone,"
zone_blocks = re.findall(r"Zone\s*,\s*([^;]+);", content, flags=re.IGNORECASE)

for block in zone_blocks:
    # ‡∏ä‡∏∑‡πà‡∏≠ zone ‡∏à‡∏∞‡∏≠‡∏¢‡∏π‡πà‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á block (‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢ comma)
    zone_name = block.strip().split(",")[0].strip().replace('"', '')
    zone_names.append(zone_name)

# ‡∏•‡∏ö‡∏ä‡∏∑‡πà‡∏≠‡∏ã‡πâ‡∏≥
zone_names = list(dict.fromkeys(zone_names))

# ‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•
print(f"‚úÖ ‡∏û‡∏ö‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {len(zone_names)} ‡πÇ‡∏ã‡∏ô:\n")
for i, z in enumerate(zone_names, 1):
    print(f"{i}. {z}")
```
